# Read CSV

```python
import pandas as pd

df = pd.read_csv('/kaggle/input/road-traffic-deaths-in-eu/road_traffic_deaths_eu.csv')
df.head()
```

# Rename columns from NUTS-0 to ISO 3166-1 alpha-3

```python
import pycountry

# first rename to alpha 2
df.rename(columns={'EL': 'GR'}, inplace=True)
df.rename(columns={'UK': 'GB'}, inplace=True)

#  and now rename to alpha 3
for col in df.columns[1:]:
    alpha_3= pycountry.countries.get(alpha_2=col).alpha_3
    df.rename(columns={f'{col}': f'{alpha_3}'}, inplace=True)
df.head()
```

# Transpose DataFrame

```python
df_transposed = df.T
df_transposed.columns = df_transposed.iloc[0].astype(int)
df_transposed = df_transposed.drop(df_transposed.index[0])
df_transposed.head()
```

# Set min and max year 

```python
years = df_transposed.columns.to_list()
print(years)
min_year = min(years)
max_year = max(years)
```

# Prepare data for map representation

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

multiplier=30
min_px=3

accidents_column_name_prefix = 'Accidents in'
norm_column_name_prefix = 'norm_'
scale_column_name_prefix = 'scale_'

scaler = MinMaxScaler() 

# iterate over years
for year in range(min_year,max_year+1):    
    # predefine column names
    accidents_column_name = f'{accidents_column_name_prefix} {year}' 
    norm_column_name = f'{norm_column_name_prefix}{year}'
    scale_column_name = f'{scale_column_name_prefix}{year}'
    
    # rename year columns
    df_transposed = df_transposed.rename(columns={year: accidents_column_name})    
   
    # normalize data
    arr_scaled = scaler.fit_transform(df_transposed[[accidents_column_name]] )
    arr_2d = np.array(arr_scaled)
    arr_1d = arr_2d.flatten()
    
    # add normalized data to dataframe
    df_transposed[norm_column_name]=arr_1d.tolist()
    
    # calculate a scale factor base on normalized data
    # this is used as bubble size
    df_transposed[scale_column_name] = round(df_transposed[norm_column_name] * multiplier + min_px)

# create column containing ISO-A3 values
df_transposed['iso_a3'] = df_transposed.index
df_transposed.head()
```

# Create dictionary of yearly dataframes

```python
df_dict = {}
for year in range(min_year,max_year+1):
    accidents_column_name = f'{accidents_column_name_prefix} {year}' 
    norm_column_name = f'{norm_column_name_prefix}{year}'
    scale_column_name = f'{scale_column_name_prefix}{year}'
    df_year = pd.DataFrame()
    df_year[str(year)] = df_transposed[accidents_column_name]
    df_year['scale'] = df_transposed[scale_column_name]
    df_year['iso_a3'] = df_transposed['iso_a3']
    df_dict[str(year)] = df_year
```

# Plot map

```python
import plotly.express as px
import plotly.graph_objects as go
fig = px.choropleth(df_transposed, color="Accidents in 2000",
                    locations="iso_a3",
                    projection="natural earth", color_continuous_scale="Reds"
                   )
fig.update_geos(scope='europe',fitbounds="locations")

# create buttons
button_list = []
for year in range (min_year, max_year + 1):
    accidents_column_name = f'{accidents_column_name_prefix} {year}' 
    button =  dict(method = "update",
                args = [
                    {'z': [ df_transposed[accidents_column_name] ] }
                ],
                label = str(year))
    button_list.append(button)
    
fig.update_layout(width=700,
                  coloraxis_colorbar_title_text = 'Road traffic deaths',
                  coloraxis_colorbar_thickness=23,
                  updatemenus=[dict(y=0.9,
                                    x=0.175,
                                    xanchor='right',
                                    yanchor='top',
                                    active=0,                                   
                                    buttons=button_list)
                              ]) 
fig.show()
```
